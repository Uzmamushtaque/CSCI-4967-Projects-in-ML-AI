## Getting started with AI Research/Projects

There is a lot going on in the field of Deep learning(DL) and AI Research and Applications. Getting started in a particular DL area can be overwhelming. Various sub-fields make up the larger area of ML and AI. Some of these are:
1. Natural Language Processing(NLP): Using AI/ML Algorithms to process and understand human text/speech. Under the Machine Learning tasks the text aspect finds more applications because speech related tasks utilize disciplines outside ML. Some of the common tasks for NLP are Sentiment Analysis, Natural Language generation, Text Summarization etc.

2. Computer Vision (CV): CV uses ML/AI models to understand and interpret digital images and videos. The result of these models can help detect objects and take some action based on that.To train a computer to understand visual data, it can be fed many labeled images and then subject to algorithms that help it identify patterns in the images.Some of the known use cases for CV are self driving cars,defect detection,medical imaging,image retrieval,image segmentation etc.

3. Machine Learning(ML): Although ML encompasses all other sub-fields mentioned in this list but from the perspective of research the core algorithms can be further studied and improved. For example optimizing learning algorithms for better performance. Ideas from Operations research can be combined to improve the existing models and algorithms. Some examples include simulations, decision analysis, stochatic programming etc.

4. Internet of Things(IoT): IoT is a network of multiple devices connected together exchanging data. This data can be analyzed for actionable results/decisions. Examples include self driving cars, Industrial IoT, healthcare IoT, Navigating apps etc.

5. Recommender Systems (RS): A RS can be based on Content-based Recommendation or even Collaborative Filtering.
Content-Based Recommendation is done by analyzing the content of all the items. For example, you can be recommended books you might like based on Natural Language Processing done on the books. On the other hand, Collaborative Filtering is done by analyzing your past reading behavior and then recommending books based on that. Both streams have their advantages and drawbacks and a lot of research is going on to improve recommendations. Hybrid recommender systems are one such area of research. Cold start problem in RS is another area of active research. Applying RS in other domains like finance education etc. is another relevant area of research.

6. Neural Networks(NN): The basic NN structure is considered to be a Universal Approximator that has the ability to learn any complex pattern. There are associated downsides to this ability namely overfitting, slow convergence etc. Experimenting with different network configurations for better overall results is another significan research area.

7. Reinforcement Learning: RL mimics the trial-and-error learning process that humans use to achieve their goals. Software actions that work towards your goal are reinforced, while actions that detract from the goal are ignored. RL algorithms excel in complex environments that have more rules and dependencies and require less human interaction. Some of the use cases are improving recommender systems. For example, an application may display ads to a user based on some demographic information. With each ad interaction, the application learns which ads to display to the user to optimize product sales.

8. Generative AI (GAI): GAI models have the ability to create novel, diverse, and useful content. GAI are revolutionizing numerous domains like music generation, healthcare,content creation, video/picture generation, fraud detection etc.

9. AI and Ethics: There is a growing concern on the ethical implications of AI and its usage. For example some of the known Language models have racial and gender bias in them. Using these as inputs to other models may lead to bias and discriminatory outcomes. One of the significant challenges in AI ethics is ensuring that AI systems are free from bias. One example of such problem could be to come up with debiasing algorithms/techniques.

## Project/Research Methodology

Once you have narrowed on a specific sub-filed of ML AI you must follow a systematic research methodology. Here's a step-by-step guide:

**Identify Project/Research Topic within the Sub-field:**

Choose a specific and well-defined topic within the AI or ML domain.
Ensure that your chosen topic is relevant and aligns with current trends or issues in the field.

**Review Existing Literature:**

Conduct a comprehensive literature review to understand the current state of research in your chosen area.
Identify gaps, challenges, or unresolved issues that your paper could address.

**Define Research Questions or Hypotheses:**

Clearly articulate the research questions or hypotheses that your paper aims to investigate or address.
Ensure that your questions are specific and can be answered through your research.

**Develop a Research Framework:**

Outline the methodology you plan to use, such as experimental design, data collection, or theoretical frameworks.
Justify your chosen approach and explain how it addresses your research questions.

**Data Collection:**

If applicable, describe the data sources, datasets, or experiments you plan to use.
Explain how the data will be collected, processed, and validated.

**Data Preprocessing and Feature Engineering:**

Detail the preprocessing steps you will undertake to clean and prepare your data.
Discuss any feature engineering techniques applied to enhance the quality of input features.

**Model Selection and Implementation:**

Specify the ML or AI models you will use and justify your choices.
Describe the implementation details, including hyperparameter settings and any modifications made to the original models.

**Evaluation Metrics:**

Define the metrics you will use to evaluate the performance of your models.
Explain why these metrics are relevant to the research questions and objectives.

**Results and Analysis:**

Present the results of your experiments or analyses.
Provide visualizations, tables, or graphs to support your findings.
Discuss the implications of your results and how they contribute to the existing body of knowledge.

**Comparison with Existing Work:**

Compare your results with those from related studies in the literature.
Highlight similarities, differences, and advancements your research brings to the field.

**Discussion and Conclusion:**

Interpret your results and discuss their significance.
Address any limitations of your study.
Summarize the key findings and their implications for the broader field.

**References:**

Ensure that you have a comprehensive list of references, following the citation style specified by the conference.

**Abstract and Introduction:**

Write a concise abstract that summarizes the key aspects of your research.
Craft an engaging introduction that provides context, introduces the problem, and outlines the paper's structure.

**Draft and Revision:**

Prepare a draft of your conference paper.
Seek feedback from colleagues, mentors, or peers, and revise your paper accordingly.

## Recources

Conducting a comprehensive literature review is crucial for understanding the current state of research in your chosen area. Here are some resources and strategies to help you with your literature review in the fields of Artificial Intelligence (AI) and Machine Learning (ML):

1. **Academic Databases:**



> a. IEEE Xplore: A digital library for electrical engineering, computer science, and electronics literature.

> b. PubMed: Focuses on biomedical and life sciences research, including machine learning applications in healthcare.

> c. ACM Digital Library: A comprehensive collection of computing and information technology literature.


2. **Google Scholar:**

> An excellent tool for finding scholarly articles, conference papers, and patents across various disciplines.

3. **ArXiv:**

> A preprint repository that hosts research papers in computer science, including AI and ML.

4. **Research Conferences and Journals:**

> Explore proceedings and publications from top conferences such as NeurIPS, ICML, CVPR, ACL, and others.
Look into reputable journals in AI and ML, such as the Journal of Machine Learning Research (JMLR) or the Journal of Artificial Intelligence Research (JAIR).

5. **Books:**

> Check relevant books written by experts in the field. Some classic texts and recent publications can provide valuable insights.

5. **Review Articles:**

> Look for review articles and survey papers that summarize and analyze the state-of-the-art in a particular subfield.

6. **Online Platforms and Blogs:**

> Follow reputable blogs and online platforms dedicated to AI and ML, such as Towards Data Science on Medium or the AI section on ArXiv Sanity Preserver.

7. **Citation Chaining:**

> Follow the citations in relevant papers to discover earlier works or foundational studies.

8. **Social Media and Online Forums:**

> Participate in discussions on platforms like Twitter, Reddit (e.g., r/MachineLearning), or specialized forums where researchers share and discuss recent advancements.

9. **Library Resources:**

> Utilize RPI university or local library resources, including interlibrary loan services, to access books and articles not available online.

## Data Sources

There are numerous data sources available for Machine Learning (ML) and Artificial Intelligence (AI) research across various domains. The choice of a data source depends on the specific focus and objectives of your research. Here are some commonly used ML and AI data sources:

**UCI Machine Learning Repository:**

A collection of databases, domain theories, and data generators commonly used by the machine learning community.

**Kaggle Datasets:**

Kaggle hosts a wide variety of datasets contributed by the community. It includes datasets for competitions as well as standalone datasets.

**Google Dataset Search:**

A tool that allows you to search for datasets across the web. It helps you discover datasets hosted by various organizations and repositories.

**AWS Public Datasets:**

Amazon Web Services provides access to various public datasets, including those relevant to machine learning and AI.

**Microsoft Research Open Data:**

A platform that hosts a variety of datasets contributed by Microsoft Research and other organizations.

**Open Data Platforms:**

Explore open data platforms from governments and institutions worldwide. For example, data.gov (U.S.), data.gov.uk (U.K.), and Eurostat (European Union) provide access to a wide range of datasets.

**ImageNet:**

A large-scale dataset for image classification, object detection, and image localization.

**COCO (Common Objects in Context):**

A dataset for object detection, segmentation, and captioning, widely used in computer vision research.

**NLP Datasets:**

For natural language processing research, datasets like the Common Crawl, SNLI (Stanford Natural Language Inference), and SQuAD (Stanford Question Answering Dataset) are commonly used.

**Healthcare Datasets:**

MIMIC-III (Medical Information Mart for Intensive Care), SEER (Surveillance, Epidemiology, and End Results), and other healthcare datasets for medical research.

**Financial Datasets:**

Yahoo Finance, Quandl, and other financial data platforms provide datasets for research in financial forecasting and analysis.

**Climate and Environmental Datasets:**

NOAA (National Oceanic and Atmospheric Administration) and NASA provide datasets related to climate, weather, and environmental research.

**Social Media Datasets:**
Datasets from platforms like Twitter, Reddit, and Facebook can be used for research in social media analytics and sentiment analysis.


Before using any dataset, it's important to review the terms of use, licensing agreements, and ethical considerations associated with the data. Additionally, consider the relevance and quality of the data for your specific research goals.
